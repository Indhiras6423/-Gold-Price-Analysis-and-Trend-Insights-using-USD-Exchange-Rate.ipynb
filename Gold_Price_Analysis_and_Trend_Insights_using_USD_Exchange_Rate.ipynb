## **Stage 2** - **Data Cleaning and Pre-processing**

```python
# Convert Date format

gold['Date'] = pd.to_datetime(gold['Date'], dayfirst=True)
usd['Date'] = pd.to_datetime(usd['Date'], dayfirst=True)



```

```python
# Sort the data
gold = gold.sort_values(by='Date')
usd = usd.sort_values(by='Date')
```

```python
# Rename columns
gold.rename(columns={
    'Close': 'Gold_Close',
    'Open': 'Gold_Open',
    'High': 'High_Gold',
    'Low': 'Low_Gold',
    'Volume': 'Volume_Gold'
}, inplace=True)

usd.rename(columns={
    'Close': 'USD_Close',
    'Open': 'USD_Open',
    'High': 'High_USD',
    'Low': 'Low_USD',
    'Volume': 'Volume_USD'
}, inplace=True)
```

```python
# Merge datasets

data = pd.merge(gold, usd, on='Date', how='inner')
```

```python
data.head(5)
```

```python
# Check data after merge
data.head()
data.info()
```

```python
# Check missing values
data.isna().sum()
```

```python
# Remove duplicates
data.drop_duplicates(inplace=True)
```

```python
# Convert Volume
data['Volume_Gold'] = data['Volume_Gold'] / 1000000
data['Volume_USD'] = data['Volume_USD'] / 1000000
```

```python
# Create new features
data['Gold_Price_Change'] = data['High_Gold'] - data['Low_Gold']
data['USD_Price_Change'] = data['High_USD'] - data['Low_USD']
```

```python
# creating new columns

# 1. Price Trend (Bullish / Bearish / Neutral)
data['Price_Trend'] = data.apply(
    lambda x: 'Bullish' if x['Gold_Close'] > x['Gold_Open']
              else ('Bearish' if x['Gold_Close'] < x['Gold_Open'] else 'Neutral'), axis=1)

# 2. Volatility Level (based on Gold_Price_Change)
q33 = data['Gold_Price_Change'].quantile(0.33)
q66 = data['Gold_Price_Change'].quantile(0.66)
data['Volatility_Level'] = data['Gold_Price_Change'].apply(
    lambda x: 'Low' if x < q33 else ('Medium' if x < q66 else 'High'))

# 3. Market Phase (based on date)
def get_phase(date):
    if date < pd.Timestamp('2020-03-01'): return 'Pre-COVID'
    elif date < pd.Timestamp('2021-01-01'): return 'COVID'
    elif date < pd.Timestamp('2022-06-01'): return 'Recovery'
    elif date < pd.Timestamp('2023-06-01'): return 'Inflation'
    else: return 'Surge'

data['Market_Phase'] = data['Date'].apply(get_phase)

# 4. Volume Category
v33 = data['Volume_Gold'].quantile(0.33)
v66 = data['Volume_Gold'].quantile(0.66)
data['Volume_Category'] = data['Volume_Gold'].apply(
    lambda x: 'Light' if x < v33 else ('Normal' if x < v66 else 'Heavy'))

```

```python
# Extract Year & Month
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
```

**Chart 8: Gold & USD Price Trends Over Time by Market Phase (Dual Axis)**

```python
fig, ax1 = plt.subplots(figsize=(12, 5))
ax2 = ax1.twinx()

ax1.plot(data['Date'], data['Gold_Close'], color='goldenrod',
         linewidth=1.2, label='Gold Close')
ax2.plot(data['Date'], data['USD_Close'], color='steelblue',
         linewidth=1, linestyle='--', alpha=0.7, label='USD Close')

# Shade by market phase
phase_colors = {
    'Pre-COVID' : '#E8F4F8',
    'COVID'     : '#FDDEDE',
    'Recovery'  : '#E8F8E8',
    'Inflation' : '#FFF3CD',
    'Surge'     : '#F3E8FF'
}
for phase, color in phase_colors.items():
    mask = data['Market_Phase'] == phase
    ax1.fill_between(data['Date'], data['Gold_Close'].min(),
                     data['Gold_Close'].max(), where=mask,
                     alpha=0.15, color=color, label=phase)

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left', fontsize=8)
plt.title("Gold vs USD Price Over Time with Market Phase Shading")
plt.tight_layout()
plt.show()
```

### **🌐 Why USD Affects Gold — 4 Core Forces**
01 — 🌍 Global Pricing in USD

Gold is priced globally in US Dollars. When the dollar strengthens, you need fewer dollars to buy the same amount of gold — making gold appear cheaper and dampening demand. When USD weakens, gold becomes more expensive globally, driving demand and price up.

02 — 🛡️ Inflation Hedge

As the USD loses purchasing power through inflation, investors turn to gold as a store of value. This relationship has held for centuries — gold preserves wealth across currency crises. During high inflation (2022–2023), gold prices surged significantly.

03 — 📊 Interest Rates

When the Fed raises rates, USD-denominated assets like bonds yield more, reducing the appeal of non-yielding gold. Lower rates have the opposite effect — pushing investors toward gold. Fed rate hikes in 2022–2023 temporarily suppressed gold despite inflation.

04 — 🏦 Safe-Haven Demand

During geopolitical crises or economic uncertainty, investors flee to BOTH gold and the USD simultaneously — temporarily breaking the inverse correlation and causing both to rise. This explains the +0.80 overall correlation in this dataset.
